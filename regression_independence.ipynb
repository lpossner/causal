```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import shapiro

N = 1000
k = 2

X = np.random.normal(size=(N, 1))
Y = k * X + np.random.normal(size=(N, 1))

k_est_xy = np.linalg.inv(X.T @ X) @ X.T @ Y
Y_est = k_est_xy * X
Y_res = Y - Y_est

k_est_yx = np.linalg.inv(Y.T @ Y) @ Y.T @ X
X_est = k_est_yx * Y
X_res = X - X_est

plt.figure()
sns.scatterplot(x=X.squeeze(), y=Y.squeeze())
sns.scatterplot(x=Y.squeeze(), y=X.squeeze())

plt.figure()
sns.scatterplot(x=X.squeeze(), y=Y.squeeze())
sns.lineplot(x=X.squeeze(), y=Y_est.squeeze(), color="red")

plt.figure()
sns.scatterplot(x=X.squeeze(), y=Y_res.squeeze())

plt.figure()
sns.scatterplot(x=Y.squeeze(), y=X.squeeze())
sns.lineplot(x=Y.squeeze(), y=X_est.squeeze(), color="red")

plt.figure()
sns.scatterplot(x=Y.squeeze(), y=X_res.squeeze())
```
